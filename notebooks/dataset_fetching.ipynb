```python
import requests
from pathlib import Path
import pandas as pd

BASE_PATH = Path("/content/trafficpulse_dataset")
SENSOR_PATH = BASE_PATH / "sensor_graph"
SENSOR_PATH.mkdir(parents=True, exist_ok=True)

sensor_url = "https://raw.githubusercontent.com/liyaguang/DCRNN/master/data/sensor_graph/graph_sensor_locations.csv"
sensor_file = SENSOR_PATH / "graph_sensor_locations.csv"

response = requests.get(sensor_url)

if response.status_code == 200:
    with open(sensor_file, "wb") as f:
        f.write(response.content)
    print("Sensor location file downloaded successfully.")
    print("Saved at:", sensor_file)
else:
    print("Failed to download sensor file.")
    print("Status code:", response.status_code)
```

```python
sensor_df = pd.read_csv(sensor_file)

print("Sensor location shape:", sensor_df.shape)
print("Columns:", sensor_df.columns.tolist())

sensor_df.head()
```

```python
traffic_sensor_ids = set(map(str, traffic_df.columns.tolist()))

print("Number of traffic sensors:", len(traffic_sensor_ids))

sensor_df.head()
```

```python
sensor_id_col = sensor_df.columns[0]

location_sensor_ids = set(sensor_df[sensor_id_col].astype(str).tolist())

matched = traffic_sensor_ids.intersection(location_sensor_ids)

print("Traffic sensors:", len(traffic_sensor_ids))
print("Location sensors:", len(location_sensor_ids))
print("Matched sensors:", len(matched))
print("Unmatched traffic sensors:", len(traffic_sensor_ids - location_sensor_ids))
```

```python
traffic_sensor_ids = set(map(str, traffic_df.columns.tolist()))

best_sensor_col = None
best_match_count = 0

for col in sensor_df.columns:
    current_ids = set(sensor_df[col].astype(str).tolist())
    match_count = len(traffic_sensor_ids.intersection(current_ids))

    if match_count > best_match_count:
        best_match_count = match_count
        best_sensor_col = col

lat_col = None
lon_col = None

for col in sensor_df.columns:
    col_lower = col.lower()
    if "lat" in col_lower:
        lat_col = col
    if "lon" in col_lower or "lng" in col_lower:
        lon_col = col

print("Detected sensor ID column:", best_sensor_col)
print("Matched sensors:", best_match_count)
print("Detected latitude column:", lat_col)
print("Detected longitude column:", lon_col)

sensor_df.head()
```

```python
if best_sensor_col is None:
    raise ValueError("Sensor ID column not detected.")

if lat_col is None or lon_col is None:
    raise ValueError("Latitude/longitude columns not detected.")

start_date = traffic_df.index.min().date()
end_date = traffic_df.index.max().date()

center_lat = sensor_df[lat_col].mean()
center_lon = sensor_df[lon_col].mean()

print("Weather start date:", start_date)
print("Weather end date:", end_date)
print("Center latitude:", center_lat)
print("Center longitude:", center_lon)
```

```python
weather_url = "https://archive-api.open-meteo.com/v1/archive"

params = {
    "latitude": center_lat,
    "longitude": center_lon,
    "start_date": str(start_date),
    "end_date": str(end_date),
    "hourly": [
        "temperature_2m",
        "relative_humidity_2m",
        "precipitation",
        "rain",
        "weather_code",
        "cloud_cover",
        "wind_speed_10m",
        "wind_gusts_10m"
    ],
    "timezone": "America/Los_Angeles"
}

response = requests.get(weather_url, params=params)

print("Status code:", response.status_code)

if response.status_code != 200:
    print(response.text)
    raise Exception("Weather data fetch failed.")

weather_json = response.json()

weather_df = pd.DataFrame(weather_json["hourly"])
weather_df["time"] = pd.to_datetime(weather_df["time"])

weather_file = WEATHER_PATH / "la_weather_metr_la_period.csv"
weather_df.to_csv(weather_file, index=False)

print("Weather dataset loaded successfully.")
print("Shape:", weather_df.shape)
print("Saved at:", weather_file)

weather_df.head()
```

```python
print("Weather start:", weather_df["time"].min())
print("Weather end:", weather_df["time"].max())
print("Columns:", weather_df.columns.tolist())

weather_df.head()
```

```python
import osmnx as ox
from shapely.geometry import box

# Bounding box around all sensors with small buffer
buffer = 0.03

north = sensor_df[lat_col].max() + buffer
south = sensor_df[lat_col].min() - buffer
east = sensor_df[lon_col].max() + buffer
west = sensor_df[lon_col].min() - buffer

print("North:", north)
print("South:", south)
print("East:", east)
print("West:", west)

polygon = box(west, south, east, north)

G = ox.graph_from_polygon(
    polygon,
    network_type="drive",
    simplify=True
)

nodes_gdf, edges_gdf = ox.graph_to_gdfs(G)

print("OSM road network loaded successfully.")
print("Nodes shape:", nodes_gdf.shape)
print("Edges shape:", edges_gdf.shape)

edges_gdf.head()
```

```python
graphml_path = OSM_PATH / "la_road_network.graphml"
edges_path = OSM_PATH / "la_road_edges.geojson"
nodes_path = OSM_PATH / "la_road_nodes.geojson"

ox.save_graphml(G, filepath=graphml_path)

edges_gdf.to_file(edges_path, driver="GeoJSON")
nodes_gdf.to_file(nodes_path, driver="GeoJSON")

print("Saved GraphML:", graphml_path)
print("Saved edges GeoJSON:", edges_path)
print("Saved nodes GeoJSON:", nodes_path)
```
